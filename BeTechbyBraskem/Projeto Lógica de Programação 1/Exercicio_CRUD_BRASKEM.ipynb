````python
def formata_telefone(numero_telefone: str) -> str:
  telefone:list = list(numero_telefone)
  telefone.insert(0, '(')
  telefone.insert(3, ')')
  telefone.insert(-4, '-')
  telefone = ''.join(telefone)
  return telefone

def formata_cep(cep: str) -> str:
  cep: list = list(cep)
  cep.insert(-3, '-')
  cep = ''.join(cep)
  return cep
````

# Função Atualiza Registro
Recebe o número de telefone do funcionário que terá o registro alterado e os dados a serem atualizados

````python
def atualiza_registro(telefone: str):
    for indice, funcionario in enumerate(funcionarios):
      if(funcionario[2] == telefone):
        print("Funcionário encontrado:")
        exibe_registro(funcionario)
        print("Abaixo, preencha os dados a serem atualizados, deixe em branco caso não deseje alterar.")
        nome = input("Nome: ")
        if nome:
            funcionarios[indice][0] = nome
        sobrenome = input("Sobrenome: ")
        if sobrenome:
            funcionarios[indice][1] = sobrenome
        telefone = input("Telefone: ")
        if telefone:
            funcionarios[indice][2] = telefone
        profissao = input("Profissão: ")
        if profissao:
            funcionarios[indice][3] = profissao
        data_nascimento = input("Data de Nascimento [dd/mm/YYYY]: ")
        if data_nascimento:
            funcionarios[indice][4] = data_nascimento
        cep = input("cep: ")
        if cep:
            funcionarios[indice][5] = cep
        funcionarios[indice] = funcionario.extend(recebe_cep_retorna_endereco(funcionarios[indice][5]))
        break
    else:
        print(f"Nenhum funcionário encontrado")
````

# Função Apaga Registro
Recebe um número de telefone, caso esse número se refira ao de algum cadastro, confirma se o usuário deseja apagar o registro desse usuário

````python
def apaga_registro(telefone: str):
    for indice, funcionario in enumerate(funcionarios):
        if funcionario[2] == telefone:
            exibe_registro(funcionario)
            confirmacao = input("Deseja realmente apagar este funcionario [s/n]?").lower()
            if confirmacao == "s":
                funcionarios.pop(indice)
                print("Usuário apagado!")
            else:
                print("Usuário não apagado!")
                
````

# Bonus 1

Quando for chamada a operação de leitura, mostrar quantos dias faltam para o aniversário pro colaborador daquele registro. Dica: use a lib `datetime`

````python
def contagem_aniversario(nome:str, data_nascimento: str):
    dia, mes, _ = data_nascimento.split('/')
    dia = int(dia)
    mes = int(mes)
    now = datetime.now()
    aniversario = datetime(now.year, mes, dia)
    if(aniversario < now):
        aniversario = datetime(now.year + 1, mes, dia)
    print(f"Falta(m) {(aniversario-now).days + 1} dia(s) para o aniversário de {nome}")
````

# Bonus 2

Receber o CEP do colaborador e armazenar seu endereco.

````python
def recebe_cep_retorna_endereco(cep:str) -> list:
  import requests
  cep = cep.replace("-", "").replace(".", "").replace(" ", "")
  if len(cep) == 8:
      link = f'https://viacep.com.br/ws/{cep}/json/'
      requisicao = requests.get(link)
      dic_requisicao = requisicao.json()
      uf = dic_requisicao['uf']
      cidade = dic_requisicao['localidade']
      bairro = dic_requisicao['bairro']
      logradouro = dic_requisicao['logradouro']
      return([uf, cidade, bairro, logradouro])
  else:
      raise Exception('CEP Inválido')
````

# Função Listar todos os funcionários

````python
def listar_funcionarios():
    for funcionario in funcionarios:
        exibe_registro(funcionario)
````

# Menu de execução do programa

A partir do menu executado no código abaixo, é possível executar todas as operações do programa.

````python
texto = '''
Bem vindo ao sistema de registro de funcionários da Braskem.\n
Veja abaixo as opções disponíveis:\n
1 - Cadastrar novo funcionário\n
2 - Buscar funcionário\n
3 - Atualizar dados de um funcionário\n
4 - Apaga o registro de um funcionário\n
5 - Listar Todos os Funcionários\n
0 - Encerrar o programa\n
Digite a opcão desejada: 
'''
````

````python
if __name__ == '__main__':
  while(True):
    opcao = int(input(texto))
    if opcao == 1:
      nome = input("Digite o nome do funcionário: ").title()
      sobrenome = input("Digite o sobrenome do funcionário: ").title()
      telefone = input("Digite o telefone do funcionário, sem caractéres especiais: ")
      telefone = telefone.replace("-", "").replace(".", "").replace(" ", "").replace("(", "").replace(")", "")
      profissao = input("Digite a profissão do funcionário: ").title()
      data_nascimento = input("Digite a data de nascimento do funcionário (dd/mm/aaaa): ")
      cep = input("Digite o cep do funcionário, sem caractéres especiais: ")
      cria_registro(nome, sobrenome, telefone, profissao, data_nascimento, cep)
    elif opcao == 2:
      tipo = int(input("Deseja buscar por:\n 1 - Nome\n 2 - Sobrenome\n 3 - Telefone\n 4 - Profissão\n 5 - Data de Nascimento\n 6 - CEP\n 7 - UF\n 8 - Cidade\n 9 - Bairro\n Digite a opção desejada:"))
      texto_busca = input("Digite o texto a ser buscado: ").title()
      busca_funcionario(tipo, texto_busca)
    elif opcao == 3:
      telefone = input("Digite o número do telefone do funcionário que deseja atualizar os dados: ")
      atualiza_registro(telefone)
    elif opcao == 4:
      telefone = input("Digite o número do telefone do funcionário que deseja apagar o registro: ")
      apaga_registro(telefone)
    elif opcao == 5:
      listar_funcionarios()
    else:
      break
````
